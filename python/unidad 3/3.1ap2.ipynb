### 2. Llamadas a funciones
Para ejecutar el código dentro de una función, debemos **llamar** a la función usando su nombre y paréntesis.

Ejemplo de llamada simple:

```python
def fun():
    print("Welcome to GFG")

fun()
```

#### Ejemplo adicional
Función de saludo:

```python
def saludo():
    print("Este es un saludo desde una función")

saludo()
```

#### Ejercicio 2.1: Doble de un número
Define una función `doblar(x)` que devuelva el doble de `x` y otra función `mostrar_doble()` que use `doblar()` e imprima el resultado para el valor 7.

```python
def doblar(x):
    return 2*x

def mostrar_doble():
    valor = 7
    print(doblar(valor))
```

```python
# Test ejercicio 2.1
assert doblar(4) == 8
import io, sys
cap = io.StringIO()
sys.stdout = cap
mostrar_doble()
sys.stdout = sys.__stdout__
assert cap.getvalue().strip() == '14'
```

#### Ejercicio 2.2: Mostrar pares hasta n
Define una función `es_par(n)` que devuelva True si `n` es par y False si es impar.
Luego crea `mostrar_pares_hasta(x)` que llame a `es_par` dentro de un bucle desde 1 hasta `x` e imprima los números pares.

```python
def es_par(n):
    if n % 2 == 0:
        return True
    else:
        return False

def mostrar_pares_hasta(x):
    for i in range(1, x+1):
        if es_par(i):
            print(i)
```

```python
# Test ejercicio 2.2
assert es_par(4) == True
assert es_par(5) == False
import io, sys
cap = io.StringIO()
sys.stdout = cap
mostrar_pares_hasta(6)
sys.stdout = sys.__stdout__
assert cap.getvalue().strip() == '2\n4\n6'
```

#### Ejercicio 2.3: Saludo condicional
Define una función `saludo_condicional(nombre, genero)` que imprima un saludo personalizado según el género:
- 'Hola Sr. {nombre}' si genero='M'
- 'Hola Sra. {nombre}' si genero='F'
- 'Hola {nombre}' si no se especifica

```python
def saludo_condicional(nombre, genero=None):
    if genero == 'M':
        print(f"Hola Sr. {nombre}")
    elif genero == 'F':
        print(f"Hola Sra. {nombre}")
    else:
        print(f"Hola {nombre}")
```

```python
# Test ejercicio 2.3
import io, sys
cap = io.StringIO()
sys.stdout = cap
saludo_condicional('Ana', 'F')
sys.stdout = sys.__stdout__
assert cap.getvalue().strip() == 'Hola Sra. Ana'

cap = io.StringIO()
sys.stdout = cap
saludo_condicional('Luis', 'M')
sys.stdout = sys.__stdout__
assert cap.getvalue().strip() == 'Hola Sr. Luis'

cap = io.StringIO()
sys.stdout = cap
saludo_condicional('Alex')
sys.stdout = sys.__stdout__
assert cap.getvalue().strip() == 'Hola Alex'
```
