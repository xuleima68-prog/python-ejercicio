### 5. Paso por valor vs referencia (mutabilidad)
En Python, los efectos de modificar un valor dentro de una función dependen del tipo de dato:
- Los tipos inmutables (enteros, strings, tuplas) **no se modifican fuera de la función**.
- Los tipos mutables (listas, diccionarios) **sí pueden modificarse dentro de la función** y esos cambios se reflejan fuera.

#### Ejemplo con lista (mutable):

```python
def modificar_lista(lst):
    lst[0] = 99

mi_lista = [1, 2, 3]
modificar_lista(mi_lista)
print(mi_lista)  # [99, 2, 3]
```

#### Ejemplo con entero (inmutable):

```python
def incrementar_num(x):
    x += 10
    return x

a = 5
b = incrementar_num(a)
print(a)  # 5
print(b)  # 15
```

#### Ejercicio 5.3: Modificar diccionario
Crea una función `actualizar_dic(dic, clave, valor)` que actualice o agregue la clave en el diccionario.

```python
def actualizar_dic(dic, clave, valor):
    dic[clave] = valor
```

```python
# Test ejercicio 5.3
mi_dic = {'a': 1}
actualizar_dic(mi_dic, 'b', 2)
assert mi_dic == {'a': 1, 'b': 2}
actualizar_dic(mi_dic, 'a', 10)
assert mi_dic['a'] == 10
```
