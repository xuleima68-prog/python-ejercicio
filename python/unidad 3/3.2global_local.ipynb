### Variables Locales y Globales en Python

En Python existen dos tipos de variables según el lugar donde se crean:

- **Variables globales:** están definidas fuera de las funciones y se pueden usar desde cualquier parte del programa.
- **Variables locales:** se crean dentro de una función y solo existen mientras la función se está ejecutando.

---
### Ejemplo 1: Variable Global
Una variable creada fuera de cualquier función puede ser utilizada dentro de ella.

```python
mensaje = "Hola desde fuera"

def mostrar_mensaje():
    print(mensaje)

mostrar_mensaje()
```

Ejemplo adicional:

```python
contador = 5

def mostrar_contador():
    print("El contador vale:", contador)

mostrar_contador()
```

---
### Ejemplo 2: Variable Local
Una variable creada dentro de una función solo existe dentro de esa función.

```python
def saludo():
    texto = "Hola desde dentro"
    print(texto)

saludo()
```

Ejemplo adicional:

```python
def doble():
    numero = 10
    return numero * 2

doble()
```

---
### Ejemplo 3: Modificar variables globales dentro de una función
Para modificar una variable global dentro de una función, es necesario usar la palabra clave `global`.

```python
contador = 0

def incrementar():
    global contador
    contador += 1

incrementar()
incrementar()
contador
```

Ejemplo adicional:

```python
estado = "apagado"

def encender():
    global estado
    estado = "encendido"

encender()
estado
```

### Ejercicio 1
Crea una función llamada `mostrar_estado()` que devuelva el contenido de la variable global `estado_actual`.

```python
estado_actual = "cargando"

def mostrar_estado():
    return estado_actual
```

```python
assert mostrar_estado() == "cargando"
```

### Ejercicio 2
Dentro de una función llamada `crear_saludo()`, crea una variable local `mensaje = "Hola!"` y devuelve su valor.

```python
mensaje = "Hola!"
def crear_saludo():
    return mensaje
```

```python
assert crear_saludo() == "Hola!"
```

### Ejercicio 3
La variable global `nivel` empieza en 1.
Crea una función `subir_nivel()` que aumente su valor en 1 usando `global`.

```python
nivel = 1

def subir_nivel():
    global nivel
    nivel +=1
    return nivel
```

```python
subir_nivel()
assert nivel == 2
```

### Ejercicio 4
Crea una función `precio_final()` con una variable local `precio = 10` que devuelva `precio * 2`.

```python
# precio global - no modificar esta línea
precio = 10

def precio_final():
    return precio * 2
```

```python
assert precio_final() == 20
```

### Ejercicio 5
Crea una variable global `modo = "normal"`.
Haz una función `cambiar_modo()` que la modifique a "turbo".
Dentro de la función crea una variable local `mensaje = "Modo cambiado"` y devuélvelo.

```python
modo = "normal"

def cambiar_modo():
    global modo
    modo = "turbo"
    mensaje = "Modo cambiado"
    return mensaje
    return modo
    
```

```python
assert cambiar_modo() == "Modo cambiado"
assert modo == "turbo"
```
