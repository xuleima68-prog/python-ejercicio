### Recursión fácil en Python

#### Ejercicio 1: Imprimir una palabra varias veces
Crea una función **repetir_palabra(palabra, n)** que devuelva la palabra repetida `n` veces en una sola cadena, usando recursión.

````python
def repetir_palabra(palabra, n):
    if n == 0:
        return ""
    else:
        return palabra + " " + repetir_palabra(palabra, n-1)
````

````python
# Tests ejercicio 1
assert repetir_palabra('Hola', 0) == ''
assert repetir_palabra('Hola', 1) == 'Hola '
assert repetir_palabra('Hola', 3) == 'Hola Hola Hola '
````

#### Ejercicio 2: Contar letras 'a'
Crea una función **contar_a(cadena)** que devuelva cuántas letras 'a' hay en una cadena usando recursión.

````python
def contar_a(cadena):
    if cadena == "":
        return 0
    cont = 0
    # la primera letra cadena[0]
    if cadena[0] == 'a' or cadena[0] == 'A':
        cont = 1
    # el resto de la cadena cadena[1:]
    return cont + contar_a(cadena[1:])
````

````python
# Tests ejercicio 2
assert contar_a('') == 0
assert contar_a('Ana') == 2
assert contar_a('Python') == 0
````

#### Ejercicio 3: Invertir cadena
Crea una función **invertir(cadena)** que devuelva la cadena al revés usando recursión. 

La última letra de la cadena se obtiene utilizando ```cadena[-1]``` y todas las letras de la cadena quitando la última letra se obtiene con ```cadena[:-1]```

````python
def invertir(cadena):
    if cadena == "": #si cadena es vacia
        return "" #devuelve cadena vacia
    return cadena[-1] + invertir(cadena[0:-1]) 
#sino devuelve el ultimo caracter mas la llamada recursiva con la cadena menos el ultimo caracter
````

````python
# Tests ejercicio 3
assert invertir('') == ''
assert invertir('Hola') == 'aloH'
assert invertir('Python') == 'nohtyP'
````

#### Ejercicio 4: Mostrar número en pantalla n veces
Crea una función **repetir_numero(num, n)** que devuelva el número repetido `n` veces como cadena separada por espacios.

````python
def repetir_numero(num, n):
    if n == 0: #mientras n sea 0
        return "" # devuelve la cadena vacia
    else: #sino
        return str(num) + " " + repetir_numero(num, n-1) 
    #que devulva el numero convertido en cadena mas un espacio mas la llamada recursiva con n-1
````

````python
# Tests ejercicio 4
assert repetir_numero(5, 0) == ''
assert repetir_numero(5, 1) == '5 '
assert repetir_numero(5, 4) == '5 5 5 5 '
````

#### Ejercicio 5: Contar espacios
Crea una función **contar_espacios(cadena)** que devuelva el número de espacios en la cadena usando recursión.

````python
def contar_espacios(cadena):
    if cadena == "":
        return 0 #mientras que cadena no tenga espacion devuelve 0
    con = 0 # contador de espacion a 0
    if cadena[0]== " ": # si contador empezando a 0 detecta un espacio
        con = 1 #suma a contador un 1
    return con + contar_espacios(cadena[1:]) # devuelve el contador mas la llamada recursiva con la cadena menos el primer caracter
````

````python
# Tests ejercicio 5
assert contar_espacios('') == 0
assert contar_espacios('Hola Mundo') == 1
assert contar_espacios('Python es divertido') == 2
````
