### 4. Usar return
La instrucción `return` finaliza la ejecución de la función y permite devolver un valor para que pueda ser usado posteriormente en otras partes del programa.

Ejemplo de función que devuelve el cuadrado de un número:

```python
def square_value(num):
    return num ** 2

print(square_value(2))  # 4
```

#### Ejemplo adicional
Función que devuelve el menor y mayor de dos números:

```python
def min_max(a, b):
    if a < b:
        return a, b
    else:
        return b, a

menor, mayor = min_max(10, 3)
print("Menor:", menor, "Mayor:", mayor)  # Menor: 3 Mayor: 10
```
