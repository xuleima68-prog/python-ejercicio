```python
#Conversión de tipo (type casting)
#“Type casting” significa convertir el valor de un tipo a otro con funciones integradas como int(), float(), str():
# Conversión básica
s = "10"    # cadena
n = int(s)  # a entero

cnt = 5
f = float(cnt)  # a flotante

age = 25
s2 = str(age)   # a cadena

print(n)   # 10
print(f)   # 5.0
print(s2)  # 25
```

```python
#Obtener el tipo de una variable
#La función integrada type() devuelve el tipo del objeto pasado:
# Definimos variables con distintos tipos
n = 42
f = 3.14
s = "Hello, World!"
li = [1, 2, 3]
d = {"key": "value"}
flag = True

# Imprimimos su tipo   
print(type(n))   # <class 'int'>
print(type(f))   # <class 'float'>
print(type(s))   # <class 'str'>
print(type(li))  # <class 'list'>
print(type(d))   # <class 'dict'>
print(type(flag))# <class 'bool'>
```

```python
#operadore aritmetico
a = 10
b = 3

print(a + b)   
print(a - b) 
print(a * b)   
print(a / b)   # división (decimal) → 3.333...
print(a // b)  # división entera → 3
print(a % b)   # módulo (resto) → 1
print(a ** b)  # potencia → 10^3 = 1000
```

```python
#operador comparacion
x = 5
y = 8

print(x == y)  # igual → False
print(x != y)  # distinto → True
print(x > y)   # mayor que → False
print(x < y)   # menor que → True
print(x >= 5)  # mayor o igual que → True
print(y <= 8)  # menor o igual que → True
```

```python
#operador logico
a = True
b = False

print(a and b)  # True si ambos son True → False
print(a or b)   # True si al menos uno es True → True
print(not a)    # Niega el valor → False
```

```python
#operador de asignacion
n = 10
n += 5   # equivale a n = n + 5 → 15
n -= 3   # equivale a n = n - 3 → 12
n *= 2   # equivale a n = n * 2 → 24
n /= 4   # equivale a n = n / 4 → 6.0
```

```python
try:
    numero = int(input("Introduce un número: "))
    if numero < 0:
        raise ValueError("Número negativo no permitido")
    resultado = 100 / numero
except ValueError as e:
    print("Error de valor:", e)
except ZeroDivisionError:
    print("No se puede dividir entre cero")
else:
    print("Resultado:", resultado)
finally:
    print("Fin del programa")

```
