## Ejercicio 1 – Clase base Persona

Crea una clase llamada `Persona` que tenga:
- Atributos: `nombre` y `edad`
- Un método `mostrar_info()` que muestre los datos por pantalla

```python
# TODO: Define la clase Persona
class Persona:

    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

# TODO: Crea un objeto Persona y prueba el método mostrar_info()

    def mostrar_info(self):
        return self.nombre, str(self.edad)
```

```python
# haz que pase estos tests
p = Persona("Ana", 20)
assert p.nombre == "Ana"
assert p.edad == 20
assert "Ana" in p.mostrar_info()
assert "20" in p.mostrar_info()
```

## Ejercicio 2 – Herencia: Alumno

Crea una clase `Alumno` que herede de `Persona` y añada:
- Un atributo `curso`
- Un método `mostrar_curso()`

```python
# TODO: Define la clase Alumno que herede de Persona
class Alumno(Persona):

    def __init__(self, nombre, edad, curso):
        super().__init__(nombre, edad)
        self.curso = curso

    def mostrar_info(self):
        return super().mostrar_info() + (self.curso,)

# TODO: Crea un objeto Alumno y muestra toda su información
```

```python
# Tests Ejercicio 2
a = Alumno("Luis", 18, "SMR")
assert a.nombre == "Luis"
assert a.edad == 18
assert a.curso == "SMR"
assert a.mostrar_curso() == "Curso: SMR"
```
