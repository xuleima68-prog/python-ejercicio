Ejercicios de diccionarios con pistas, sin soluciones. Incluye ejercicios de básico a avanzado.

### Instrucciones:
- Completa las variables `None` o vacías según la pista.
- Ejecuta cada celda, los `assert` verifican la estructura.
- Practica operaciones básicas y avanzadas de diccionarios.

## Ejercicios Básicos

```python
# Ejercicio 1: Crear un diccionario con edades de 3 personas
edades = {'Ana': 20, 'Luis': 22, 'Marta': 19}  # Pista: {'Ana': 20, 'Luis': 22, 'Marta': 19}
assert isinstance(edades, dict) and len(edades)==3, "Debe ser un diccionario con 3 elementos"
```

```python
# Ejercicio 2: Acceder a la edad de 'Ana'
edad_ana = edades.get('Ana')  # Pista: edades['Ana'] o edades.get('Ana')
assert edad_ana is not None, "Debes asignar la edad de Ana"
```

```python
# Ejercicio 3: Añadir a 'Carlos' con edad 25
edades['Carlos'] = 25  # Pista: asigna un valor a la clave 'Carlos'
assert 'Carlos' in edades, "Debes añadir a Carlos"
```

```python
# Ejercicio 4: Modificar la edad de 'Ana' a 21
edades['Ana'] = 21  # Pista: asigna el nuevo valor a la clave existente
assert edades['Ana'] == 21, "La edad de Ana debe ser 21"
```

```python
# Ejercicio 5: Eliminar la clave 'Luis'
# Pista: puedes usar del edades['Luis'] o edades.pop('Luis')
edades.pop('Luis')
assert 'Luis' not in edades, "Luis debe ser eliminado"
```

## Ejercicios Intermedios

```python
# Ejercicio 6: Crear un diccionario de productos y precios
productos = {'Leche':1.5, 'Pan':0.8, 'Huevos':0.2}  # Pista: {'Leche':1.5, 'Pan':0.8, 'Huevos':0.2}
assert isinstance(productos, dict) and len(productos)==3, "Debe ser un diccionario de 3 productos"
```

```python
# Ejercicio 7: Sumar el precio de todos los productos
total = sum(productos.values())  # Pista: sum(productos.values())
assert total is not None, "Debes calcular la suma total"
```

```python
# Ejercicio 8: Comprobar si 'Pan' está en productos
existe = 'Pan' in productos  # Pista: 'Pan' in productos
assert existe is not None, "Debes comprobar la existencia"
```

```python
# Ejercicio 9: Obtener listas de claves y valores
claves = [list(productos.keys())]  # Pista: list(productos.keys())
valores = [list(productos.values())]  # Pista: list(productos.values())
assert isinstance(claves, list) and isinstance(valores, list), "Debe ser listas"
```

```python
# Ejercicio 10: Recorrer diccionario e imprimir claves y valores
# Pista: for clave, valor in productos.items(): ...
for clave, valor in productos.items():
    print(f"{clave}: {valor}")
```

## Ejercicios Avanzados

```python
# Ejercicio 11: Diccionario de estudiantes y notas (listas como valores)
notas = {'Ana':[7,8,9], 'Luis':[6,7,8], 'Marta':[9,9,10]}  # Pista: {'Ana':[7,8,9], 'Luis':[6,7,8], 'Marta':[9,9,10]}
assert isinstance(notas, dict) and len(notas)==3, "Debe ser diccionario con 3 estudiantes"
```

```python
# Ejercicio 12: Calcular promedio de Ana
promedio_ana = sum(notas['Ana'])/len(notas['Ana'])  # Pista: sum(notas['Ana'])/len(notas['Ana'])
assert promedio_ana is not None, "Debes calcular el promedio"
```

```python
# Ejercicio 13: Filtrar productos con precio > 1
productos_caros = [p for p,v in productos.items() if v>1]  # Pista: [p for p,v in productos.items() if v>1]
assert isinstance(productos_caros, list), "Debe ser lista"
```

```python
# Ejercicio 14: Actualizar nota de Luis a 10
notas_actualizadas = { "Luis": [10] }  # Pista: copia notas y reemplaza la lista de Luis
assert isinstance(notas_actualizadas, dict), "Debe ser diccionario"
```

```python
# Ejercicio 15: Crear diccionario con claves numéricas
numeros = {1:'uno', 2:'dos', 3:'tres'}  # Pista: {1:'uno', 2:'dos', 3:'tres'}
assert isinstance(numeros, dict) and len(numeros)==3, "Debe ser diccionario con 3 elementos"
```

```python
# Ejercicio 16: Invertir claves y valores
invertido = {v:k for k,v in numeros.items()}  # Pista: {v:k for k,v in numeros.items()}
print(invertido)
assert isinstance(invertido, dict) and len(invertido)==3, "Debe ser diccionario invertido"
```

```python
# Ejercicio 17: Eliminar estudiantes con promedio < 8
notas_altas = {clave:valor for clave, valor in notas.items() if sum(valor)/len(valor) >=8}  # Pista: filtra notas usando comprensión de diccionario
assert isinstance(notas_altas, dict), "Debe ser diccionario"
```

```python
# Ejercicio 18: Contar cuántos productos tienen precio < 1
contador = sum(1 for v in productos.values() if v<1)  # Pista: sum(1 for v in productos.values() if v<1)
print(contador)
assert contador is not None, "Debes contar correctamente"
```

```python
# Ejercicio 19: Actualizar todos los precios subiéndolos 10%
productos_nuevos = {clave:valor*1.1 for clave, valor in productos.items()}  # Pista: {k:v*1.1 for k,v in productos.items()}
assert isinstance(productos_nuevos, dict), "Debe ser diccionario"
```

```python
# Ejercicio 20: Merge de dos diccionarios
d1 = {'a':1,'b':2}
d2 = {'b':3,'c':4}
d_merged = {**d1, **d2}  # Pista: {**d1, **d2}
assert isinstance(d_merged, dict) and len(d_merged)==3, "Debe ser diccionario mergeado"
```
