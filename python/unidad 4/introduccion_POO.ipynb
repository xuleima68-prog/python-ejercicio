## Ejercicio 1 ‚Äì Clase Camiseta

Crea la clase `Camiseta` con atributos: `talla`, `color`, `precio`. A√±ade un m√©todo `descripcion()` que devuelva un texto con los datos de la camiseta.

```python
# Escribe tu c√≥digo aqu√≠
class Camiseta:
    def __init__(self, talla, color, precio,  material=None):
        self.talla = talla
        self.color = color
        self.precio = precio
        self.material = material
        pass
    def descripcion(self):
        if (self.material is None):
            return f"Camiseta {self.color}, talla {self.talla}, {self.precio}‚Ç¨"
        else:
            return f"Camiseta {self.color}, talla {self.talla}, {self.precio}‚Ç¨, material {self.material}"
    
    def tenir(self, color_tinte):
        self.color = color_tinte
    
    def rebajar(self, porcentaje):
        self.precio = self.precio *((100 - porcentaje)/100)
```

```python
c1 = Camiseta("M", "roja", 15)
assert c1.descripcion() == "Camiseta roja, talla M, 15‚Ç¨"

camiseta2 = Camiseta("S", "rosa", 10)
assert camiseta2.descripcion() == "Camiseta rosa, talla S, 10‚Ç¨"

camiseta2.tenir("blanco")
assert camiseta2.color == "blanco"
assert camiseta2.descripcion() == "Camiseta blanco, talla S, 10‚Ç¨"

camiseta2.rebajar(20)
assert camiseta2.precio == 8

camiseta3 = Camiseta("L", "verde", 20, "lino")
assert camiseta3.descripcion() == "Camiseta verde, talla L, 20‚Ç¨, material lino"
```

## Ejercicio 2 ‚Äì Clase Viaje

Crea la clase `Viaje` con atributos: `destino`, `dias`, `precio`. A√±ade un m√©todo `coste_diario()` que devuelva el precio dividido entre los d√≠as.

```python
# Escribe tu c√≥digo aqu√≠
class Viaje:
    def __init__(self, destino, dias, precio):
        self.destino = destino
        self.dias = dias
        self.precio = precio
        
    def coste_diario(self):
        return self.precio / self.dias
```

```python
v = Viaje("Roma", 4, 400)
assert v.coste_diario() == 100
```

## Ejercicio 3 ‚Äì Clase Jugador

Crea la clase `Jugador` con atributos: `nombre`, `dorsal`, `goles`. A√±ade un m√©todo `marcar_gol()` que sume 1 gol al atributo `goles`.

```python
# Escribe tu c√≥digo aqu√≠
class Jugador:
    def __init__(self, nombre, dorsal, goles=0):
        self.nombre = nombre
        self.dorsal = dorsal
        self.goles = goles
        
    def separar(self):
        if (self.goles is None):
            return f"{self.nombre}, {self.dorsal}"
        else:
            return f"{self.nombre}, {self.dorsal}, {self.goles}"

    def marcar_gol(self):
        self.goles += 1
        pass
```

```python
j = Jugador("Alex", 9)
j.marcar_gol()
j.marcar_gol()
assert j.goles == 2
```

## Ejercicio 4 ‚Äì Clase Cancion

Crea la clase `Cancion` con atributos: `titulo`, `artista`, `duracion`. A√±ade un m√©todo `es_larga()` que devuelva True si la duraci√≥n es mayor a 5 minutos.

```python
# Escribe tu c√≥digo aqu√≠
class Cancion:
    def __init__(self, titulo, artista, duracion):
        self.titulo = titulo
        self.artista = artista
        self.duracion = duracion

    def es_larga(self):
        if self.duracion > 5:
            return True
        else:
            return False
```

```python
c = Cancion("Epic Song", "DJ Max", 6)
assert c.es_larga() is True
```

## Ejercicio 5 ‚Äì Clase Mascota

Crea la clase `Mascota` con atributos: `nombre`, `tipo`, `energia`. A√±ade m√©todos: `jugar()` que reste 10 a energ√≠a y `dormir()` que sume 20 a energ√≠a.

```python
# Escribe tu c√≥digo aqu√≠. La energ√≠a se inicializa en 50
class Mascota:
    def __init__(self, nombre, tipo, energia = 50):
        self.nombre = nombre
        self.tipo = tipo
        self.energia = energia

    def jugar(self):
        self.energia -= 10

    def dormir(self):
        self.energia += 20
        
```

```python
m = Mascota("Luna", "perro")
m.jugar()
m.dormir()
assert m.energia == 60
```
