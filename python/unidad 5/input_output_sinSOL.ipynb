# Entrada y salida en Python

## Ejercicio 1 – Entrada básica

Crea una función `saludar(nombre)` que devuelva el texto utilizando f-String:

`Hola <nombre>`

```python
def saludar(nombre):
    return f"Hola {nombre}"

assert saludar("Ana") == "Hola Ana"
assert saludar("Luis") == "Hola Luis"
```

## Ejercicio 2 – Conversión de tipos

Crea una función `sumar(a, b)` que reciba dos cadenas numéricas y devuelva su suma como entero.

```python
def sumar(a, b):
    return int(a) + int(b)

assert sumar("3", "5") == 8
assert sumar("10", "2") == 12
```

## Ejercicio 3 – Formateo de salida

Crea una función `info_producto(nombre, precio)` que devuelva (utlizando f-string):

`El producto X cuesta Y euros`

```python
def info_producto(nombre, precio):
    return f"El producto {nombre} cuesta {precio} euros"

assert info_producto("Pan", 0.8) == "El producto Pan cuesta 0.8 euros"
assert info_producto("Leche", 1.5) == "El producto Leche cuesta 1.5 euros"
```

## Ejercicio 4 – Múltiples datos

Crea una función `datos_persona(nombre, edad, ciudad)` que devuelva:

`Nombre: X | Edad: Y | Ciudad: Z`

```python
def datos_persona(nombre, edad, ciudad):
    return f"Nombre: {nombre} | Edad: {edad} | Ciudad: {ciudad}"

assert datos_persona("Ana", 18, "Madrid") == "Nombre: Ana | Edad: 18 | Ciudad: Madrid"
```

## Ejercicio 5 – Escritura en archivo

Crea una función `guardar_nombre(nombre)` que guarde el nombre en `usuarios.txt`.

```python
def guardar_nombre(nombre):
    with open("usuarios.txt", "w") as f:
        return f.write(nombre)

guardar_nombre("Ana")
with open("usuarios.txt", "r") as f:
    assert f.read() == "Ana"
```

## Ejercicio 6 – Lectura de archivo

Crea una función `leer_usuarios()` que devuelva el contenido de `usuarios.txt`.

```python
def leer_usuarios():
    with open("usuarios.txt", "r") as f:
        return f.read()

assert leer_usuarios() == "Ana"
```

## Ejercicio 7 – Añadir datos a un archivo

Crea una función `añadir_usuario(nombre)` que añada un nombre nuevo a `usuarios.txt`.

```python
def añadir_usuario(nombre):
    with open("usuarios.txt", "a") as f:
        f.write("\n" + nombre)

añadir_usuario("Luis")
with open("usuarios.txt", "r") as f:
    contenido = f.read()

assert "Ana" in contenido
assert "Luis" in contenido
```

## Ejercicio 8 – Uso de librería os

Crea una función `archivo_existe(nombre_archivo)` que devuelva `True` o `False`.

```python
import os

def archivo_existe(nombre_archivo):
    return os.path.exists(nombre_archivo)

assert archivo_existe("usuarios.txt") is True
assert archivo_existe("no_existe.txt") is False
```

## Ejercicio 9 – Registro de notas

Crea una función `guardar_nota(nombre, nota)` que guarde en `notas.txt` el formato:

`nombre,nota`

```python
def guardar_nota(nombre, nota):
    with open("notas.txt", "a") as f:
        f.write(f"{nombre},{nota}\n")

guardar_nota("Ana", 7.5)
guardar_nota("Luis", 8.0)

with open("notas.txt", "r") as f:
    texto = f.read()

assert "Ana,7.5" in texto
assert "Luis,8.0" in texto
```

## Ejercicio 10 – Mini ejercicio integrador

Crea una función `resumen_alumno(nombre, edad)` que:
- Devuelva un texto formateado
- Guarde el resumen en `resumen.txt`

```python
def resumen_alumno(nombre, edad):
    texto = f"Alumno: {nombre} - Edad: {edad}"
    with open("resumen.txt", "w") as f:
        f.write(texto)
    return texto

resultado = resumen_alumno("Ana", 18)
assert resultado == "Alumno: Ana - Edad: 18"

with open("resumen.txt", "r") as f:
    assert f.read() == "Alumno: Ana - Edad: 18"
```
